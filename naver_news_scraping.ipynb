```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin


# print(type(res))
# print('Response Header: ', res.headers)
# print('Request Header: ', res.request.headers)
```

```python
def print_news_info(category):
    category_dict = {'정치': 100, '경제': 101, '사회': 102, '생활/문화': 103, 'IT/과학': 105, '세계': 104}

    # dictioary get의 기능 -> Key에 대응되는 Value를 돌려준다.
    # get을 몰라서 한 삽질 -> for문 돌려서 조건에 맞으면 url에 join 시킴
    code = category_dict.get(category)

    if code:
        url = f'https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1={code}'

    req_header_dict = {
        # 요청 헤더 : 브라우저 정보
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36'
    }

    res = requests.get(url, headers=req_header_dict)

    print(url)

    print(res.status_code, res.ok)

    if res.ok:
        html = res.text
        soup = BeautifulSoup(html, 'html.parser')
        # href 속성 값에 read.naver 문자열이 포함된 a tag 선택
        atag_list = soup.select("a[href*='read.naver']")
        print(len(atag_list), atag_list)

        for idx, atag in enumerate(atag_list, 1):
            # <a>뉴스제목</a>
            title = atag.text.strip()
            # <a href="url">
            news_link = urljoin(url, atag['href'])
            print(title, news_link)
```

```python
print_news_info('생활/문화')
```
